```python
import scipy.spatial
import pandas as pd
```

```python
filename = 'bag.csv'
folder = 'C:/Users/jullian/Desktop/Jullian/Jullian/Data/Qgis/open data amsterdam/bag/'
path = folder + filename
feitgebruik = pd.read_csv(path,index_col = 0)

feitgebruik['PC4'] = feitgebruik['Postcode'].str.slice(0,4) # change 6-digit postcode to 4-digit
feitgebruik['Feitelijk_gebruik'].fillna('missing',inplace = True) # if building use function is 'NaN' fill in missing
```

```python
# this function computes the land_use_distribution for each PC4
# result is a dataframe showing the land_use_distribution for each PC4
def land_use_distribution(bag_df):
    df = pd.DataFrame(index = feitgebruik['Feitelijk_gebruik'].unique())
    for i in feitgebruik['PC4'].unique():
        index = feitgebruik['PC4'] == i
        selection = feitgebruik[index]
        land_use_distribution = pd.DataFrame(selection['Feitelijk_gebruik'].value_counts(normalize = True))
        land_use_distribution.columns = {i}
        df = df.merge(land_use_distribution,left_index = True,right_index = True,how = 'left',sort = False)
        df[i].fillna(0,inplace = True)
    return df
```

```python
PC4 = land_use_distribution(feitgebruik)
#filename = 'Feitgebruik_per_PC4.csv'
#folder = 'C:/Users/jullian/Desktop/Jullian/Jullian/Data/SimPark/Jullian/'
#path = folder + filename
#PC4 = pd.read_csv(path,index_col = 0) 
```

```python
PC4_t = PC4.transpose()
select = pd.DataFrame(PC4['1022']).transpose() # 1022 is the postcode with the NZL

PC4_t.describe()
# there are several more metrics available in the scipy package.
# for the first try i decided to keep it simple however.
euclidean_distance = scipy.spatial.distance.cdist(PC4_t, select, metric='euclidean')
cosine_distance = scipy.spatial.distance.cdist(PC4_t, select, metric='cosine')

PC4_t['euclidean_distance'] = euclidean_distance
PC4_t['cosine_distance'] = cosine_distance
PC4_t.sort_values(by = ['euclidean_distance'])
PC4_t.to_csv('PC4.csv')
```
